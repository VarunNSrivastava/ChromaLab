```python
import csv
import numpy as np
from spectra import Spectra, Pigment
import matplotlib.pyplot as plt
import numpy.typing as npt
import torch as th
```

```python
from observer import Observer, Cone
from inks import load_neugebauer, InkGamut, Neugebauer
```

```python
trichromat = Observer.trichromat()
tetrachromat = Observer.tetrachromat()
```

```python
cmy_primaries_dict = {}
wavelengths1 = np.arange(390, 701, 1)
wavelengths10 = np.arange(400, 701, 10)

primary_fns = [
    "000",
    "001",
    "010",
    "100",
    "011",
    "110",
    "101",
    "111",
]

for fn in primary_fns:
    with open(f'PrintColors/{fn}.csv') as csvf:
        spamreader = csv.reader(csvf, delimiter=';')
        for i, row in enumerate(spamreader):
            if i == 4:
                color_data = np.array(row[33:],dtype=float)
                spectra = Spectra(data=color_data, wavelengths=wavelengths10)
                cmy_primaries_dict[fn] = spectra
```

```python
all_inks = {}
fig, axc = plt.subplots()
with open(f'Inks_all.csv') as csvf:
    spamreader = csv.reader(csvf, delimiter=';')
    for i, row in enumerate(spamreader):
        if i >= 4:
            name = row[4]
            color_data = np.array(row[33:],dtype=float)
            spectra = Spectra(data=color_data, wavelengths=wavelengths10)
            spectra.plot(ax=axc)
            
            all_inks[name] = spectra
            
plt.show()
```

```python
len(all_inks)
```

```python
paper = cmy_primaries_dict["000"]
cyan = cmy_primaries_dict["100"]
magenta = cmy_primaries_dict["010"]
yellow = cmy_primaries_dict["001"]
```

```python
widths = []
for name, spectra in all_inks.items():
    gamut = InkGamut([cyan, magenta, yellow, spectra], paper)
    width = gamut.get_width(tetrachromat, stepsize=0.05)
    widths.append((width, name))

widths.sort(reverse=True)
```

```python
peach_haze = all_inks["Peach haze D"]
```

```python
yellow_widths = []
for name, spectra in all_inks.items():
    gamut = InkGamut([cyan, magenta, spectra, peach_haze], paper)
    width = gamut.get_width(tetrachromat, stepsize=0.05)
    yellow_widths.append((width, name))

yellow_widths.sort(reverse=True)
```
